```python
def filter_nouns(corpus):
  filtered = []
  for line in corpus:
    opt = nltk.tag.pos_tag(line.split())
    eopt = [word for word,tag in opt if tag != 'NNP' and tag != 'NNPS']
    filtered.append(' '.join(eopt))

  return filtered
```
